```python
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Features to compare across cities
features_to_test = ['revenue_per_customer', 'cost', 'customer_count', 'cases']

# Store results
anova_results = {}
tukey_results = {}

for feature in features_to_test:
    # Group values by city
    groups = [group[feature].values for name, group in df.groupby('city')]
    
    # Run one-way ANOVA
    f_stat, p_val = f_oneway(*groups)
    anova_results[feature] = {'F-statistic': f_stat, 'p-value': p_val}
    
    # Run Tukey HSD for post-hoc pairwise testing
    tukey = pairwise_tukeyhsd(endog=df[feature], groups=df['city'], alpha=0.05)
    tukey_results[feature] = tukey.summary()

# View ANOVA results
for feature, result in anova_results.items():
    print(f"\n=== {feature.upper()} ===")
    print(f"F-statistic: {result['F-statistic']:.4f}, p-value: {result['p-value']:.4e}")
    if result['p-value'] < 0.05:
        print("=> Statistically significant difference across cities.")
    else:
        print("=> No significant difference across cities.")

# Optional: View a specific Tukey result
print("\n\nTukey HSD for COST:")
print(tukey_results['cost'])
```

First going to create some functions to help us with this

Absolutely — let’s break down each helper function so you’re crystal clear on what you just engineered.

⸻
 Helper Function Breakdown

⸻

1. add_time_features(df)

df['week_num'] = df['week'].dt.isocalendar().week
df['month'] = df['week'].dt.month

What it does:
	•	Adds week_num (1 to 52) — helpful for capturing seasonal effects
	•	Adds month (1 to 12) — captures monthly patterns like holiday surges, seasonal cycles

Why it matters:
Some businesses are cyclical (e.g., summer spikes, end-of-quarter dips). Time-aware features help models learn those patterns.

⸻

2. add_lag_features(df, lag_vars=['revenue'], lags=[1, 2])

for var in lag_vars:
    for lag in lags:
        df[f'{var}_lag{lag}'] = df.groupby(['cuisine', 'product_type'])[var].shift(lag)

What it does:
	•	For each var (e.g., revenue), creates:
	•	revenue_lag1: value from 1 week ago
	•	revenue_lag2: value from 2 weeks ago
	•	Uses .groupby(['cuisine', 'product_type']) so comparisons are meaningful within a product type

Why it matters:
Revenue today is highly correlated with recent weeks. Lags give the model memory.

⸻

3. add_rolling_features(df, roll_vars=['revenue', 'cost'], windows=[3])

for var in roll_vars:
    for window in windows:
        df[f'{var}_roll{window}'] = df.groupby(['cuisine', 'product_type'])[var].shift(1).rolling(window=window).mean()

What it does:
	•	Calculates 3-week moving averages of revenue and cost per product/cuisine combo
	•	Uses .shift(1) so we don’t peek into the future

Why it matters:
Smooths out noise. Helps models understand longer-term trends (e.g., building demand or cost creep).

⸻

4. pd.get_dummies(...)

df = pd.get_dummies(df, columns=['cuisine', 'product_type'], drop_first=True)

What it does:
	•	Turns categorical text fields (cuisine, product_type) into numeric one-hot encoded columns
	•	drop_first=True prevents multicollinearity (avoids dummy trap)

Why it matters:
Models can’t work directly with text — encoding lets them learn effects by category (e.g., “Asian cuisine tends to sell more”).

⸻

Final Cleanup:

df = df.dropna().reset_index(drop=True)

Why:
	•	Lag and rolling operations naturally create NaN at the start — we drop those to avoid training issues

⸻
 Summary:

These functions add memory, seasonality, and categorical structure — giving your model the tools to learn complex, time-aware patterns.

```python
def add_time_features(df):
    df['week_num'] = df['week'].dt.isocalendar().week
    df['month'] = df['week'].dt.month
    return df

def add_lag_features(df, lag_vars=['revenue'], lags=[1, 2]):
    for var in lag_vars:
        for lag in lags:
            df[f'{var}_lag{lag}'] = df.groupby(['cuisine', 'product_type'])[var].shift(lag)
    return df

def add_rolling_features(df, roll_vars=['revenue', 'cost'], windows=[3]):
    for var in roll_vars:
        for window in windows:
            df[f'{var}_roll{window}'] = df.groupby(['cuisine', 'product_type'])[var].shift(1).rolling(window=window).mean()
    return df
```

```python
# Reuse train_sets dict
engineered_train_sets = {}

for city, df in train_sets.items():
    df = df.copy()

    # Step 1: Time-based features
    df = add_time_features(df)

    # Step 2: Lag features
    df = add_lag_features(df, lag_vars=['revenue'], lags=[1, 2])

    # Step 3: Rolling features
    df = add_rolling_features(df, roll_vars=['revenue', 'cost'], windows=[3])

    # Step 4: One-hot encode categorical variables
    df = pd.get_dummies(df, columns=['cuisine', 'product_type'], drop_first=True)

    # Drop rows with NA after lag/rolling
    df = df.dropna().reset_index(drop=True)

    engineered_train_sets[city] = df
    print(f"{city}: Engineered train shape = {df.shape}")
```

```python
# Store results
scores = {
    'RandomForest': {},
    'XGBoost': {}
}
models = {
    'RandomForest': {},
    'XGBoost': {}
}

for city in engineered_train_sets.keys():
    # Prep training data
    train = engineered_train_sets[city]
    X_train = train.drop(columns=['revenue', 'week', 'city'])
    y_train = train['revenue']

    # Prepare and engineer test set
    test = test_sets[city].copy()
    test['week_num'] = test['week'].dt.isocalendar().week
    test['month'] = test['week'].dt.month
    for lag in [1, 2]:
        test[f'revenue_lag{lag}'] = test.groupby(['cuisine', 'product_type'])['revenue'].shift(lag)
    for window in [3]:
        test[f'revenue_roll{window}'] = test.groupby(['cuisine', 'product_type'])['revenue'].shift(1).rolling(window).mean()
        test[f'cost_roll{window}'] = test.groupby(['cuisine', 'product_type'])['cost'].shift(1).rolling(window).mean()

    test = pd.get_dummies(test, columns=['cuisine', 'product_type'], drop_first=True)
    test = test.dropna().reset_index(drop=True)

    X_test = test.drop(columns=['revenue', 'week', 'city'])
    y_test = test['revenue']
    X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

    # ---------------- Random Forest ----------------
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train)
    y_rf_pred = rf_model.predict(X_test)
    models['RandomForest'][city] = rf_model

    scores['RandomForest'][city] = {
        'RMSE': np.sqrt(mean_squared_error(y_test, y_rf_pred)),
        'MAE': mean_absolute_error(y_test, y_rf_pred),
        'R2': r2_score(y_test, y_rf_pred)
    }

    # ---------------- XGBoost ----------------
    xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
    xgb_model.fit(X_train, y_train)
    y_xgb_pred = xgb_model.predict(X_test)
    models['XGBoost'][city] = xgb_model

    scores['XGBoost'][city] = {
        'RMSE': np.sqrt(mean_squared_error(y_test, y_xgb_pred)),
        'MAE': mean_absolute_error(y_test, y_xgb_pred),
        'R2': r2_score(y_test, y_xgb_pred)
    }

    # Print comparison
    print(f"\n{city}")
    print("Random Forest  ->  RMSE: {:.2f}, MAE: {:.2f}, R²: {:.3f}".format(
        scores['RandomForest'][city]['RMSE'],
        scores['RandomForest'][city]['MAE'],
        scores['RandomForest'][city]['R2']
    ))
    print("XGBoost        ->  RMSE: {:.2f}, MAE: {:.2f}, R²: {:.3f}".format(
        scores['XGBoost'][city]['RMSE'],
        scores['XGBoost'][city]['MAE'],
        scores['XGBoost'][city]['R2']
    ))
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Define tuned XGBoost model
def get_tuned_xgb_model():
    return XGBRegressor(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )

# Evaluation results storage
xgb_metrics_tuned = {}

# Loop over each city
for city in engineered_train_sets:
    train_df = engineered_train_sets[city].copy()
    test_df = test_sets[city].copy()

    # Feature engineering on test set
    test_df = add_time_features(test_df)
    test_df = add_lag_features(test_df)
    test_df = add_rolling_features(test_df)
    test_df = pd.get_dummies(test_df, columns=['cuisine', 'product_type'], drop_first=True)
    test_df = test_df.dropna().reset_index(drop=True)

    # Skip if not enough test data left
    if len(test_df) < 30:
        print(f"Skipping {city}: Not enough test data after feature generation.")
        continue

    # Align columns
    X_train = train_df.drop(columns=['revenue', 'week', 'city'])
    y_train = train_df['revenue']
    X_test = test_df.drop(columns=['revenue', 'week', 'city'])
    y_test = test_df['revenue']
    X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

    # Train and predict
    model = get_tuned_xgb_model()
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    # Calculate metrics
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    mae = mean_absolute_error(y_test, preds)
    r2 = r2_score(y_test, preds)

    xgb_metrics_tuned[city] = {
        'RMSE': round(rmse, 2),
        'MAE': round(mae, 2),
        'R2': round(r2, 3)
    }

# Display results
for city, metrics in xgb_metrics_tuned.items():
    print(f"\n{city}")
    for metric, value in metrics.items():
        print(f"{metric}: {value}")
```
